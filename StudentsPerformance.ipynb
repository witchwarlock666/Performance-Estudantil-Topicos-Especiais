```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='gender', hue='gender', data=df, palette="Set1")
```

```python
sns.countplot(x='race/ethnicity', hue='race/ethnicity', data=df, palette="Set1")
```

```python
plt.figure(figsize=(12, 4))
sns.countplot(x='parental level of education', hue='parental level of education', data=df, palette="Set1", )
```

```python
sns.countplot(x='test preparation course', hue='test preparation course', data=df, palette="Set1", )
```

```python
cat = df.select_dtypes(include=['object']).columns.tolist()
for col in cat:
    print(col)
    print(df[col].unique())
```

```python
df['gender'] = df['gender'].replace({'female': 0, 'male': 1})
df['parental level of education'] = df['parental level of education'].replace({'some high school': 0, 'high school': 1, 'associate\'s degree': 2, 'some college': 3, 'bachelor\'s degree': 4, 'master\'s degree': 5})
df['race/ethnicity'] = df['race/ethnicity'].replace({'group A': 0, 'group B': 1, 'group C': 2, 'group D': 3, 'group E': 4})
df['lunch'] = df['lunch'].replace({'standard': 0, 'free/reduced': 1})
df['test preparation course'] = df['test preparation course'].replace({'none': 0, 'completed': 1})

df.head()
```

```python
sns.scatterplot(x=df['test preparation course'], y=df['math score'])
plt.show()
sns.scatterplot(x=df['test preparation course'], y=df['reading score'])
plt.show()
sns.scatterplot(x=df['test preparation course'], y=df['writing score'])
plt.show()

sns.scatterplot(x=df['gender'], y=df['math score'])
plt.show()
sns.scatterplot(x=df['gender'], y=df['reading score'])
plt.show()
sns.scatterplot(x=df['gender'], y=df['writing score'])
plt.show()

sns.scatterplot(x=df['race/ethnicity'], y=df['math score'])
plt.show()
sns.scatterplot(x=df['race/ethnicity'], y=df['reading score'])
plt.show()
sns.scatterplot(x=df['race/ethnicity'], y=df['writing score'])
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```
